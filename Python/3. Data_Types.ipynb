###### _**2. Strings -**_

```python
Example_string = "String"

print(Example_string)       # Printing the value of variable
print(Example_string[2])    # Printing value at index 2
print(Example_string[2:4])  # Slicing the variable
```

###### _3. List -_

```python
Example_list = [11, 22, 33, 44, 55]

print(Example_list)         # Print the list
print(Example_list[2])      # Print list index 2
print(Example_list[1:4])    # Slicing the list
```

##### _4. Tuple -_

```python
Example_tuple = (45, 'Python', 78.08)   # Creating a tuple

print(Example_tuple)                    # Print the tuple
print(Example_tuple[2])                 # Print tuple index 2
print(Example_tuple[0:1])               # Slicing the tuple
```

##### _5. Swapping two numbers using tuple_

```python
num1 = 67                # 1st number
num2 = 78                # 2nd number

num1, num2 = num2, num1  # swap the values
print("num1 =", num1)    # printing 1st number after swapping
print("num2 =", num2)    # printing 2nd number after swapping
```

##### _6. Implemention of a Python dictionary_

```python
MyDict = {}         # Empty dictionary
MyDict['one'] = "This is Python"
MyDict[2] = "This is Python again"
Dict = {'Roll no.' : 1, 'Name' : 'Smith', 'Age' : 23}  # Creating a dictionary
print (MyDict[ 'one' ])   # Printing the value of the key 'one'
print (MyDict[ 2 ])       # Printing the value of the key 2
print (Dict[ 'Age' ])     # Printing the value of the key 'Age'
print (Dict.keys())       # Printing the keys
print (Dict.values())     # Printing the values
```

##### _7._ _Sets_

```python
MySet = {11, 33}   # Creating a set
print(MySet)       # Printing the set
MySet.add(22)      # Adding an element to the set
print(MySet)       # Printing the set
MySet.remove(11)   # Removing an element from the set
print(MySet)       # Printing the set
```
