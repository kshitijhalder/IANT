Diffie-Hellman Algorithm

```python
# Power function to return value of a^b mod P
def power(a, b, p):
    if b == 1:
        return a
    else:
        return pow(a, b) % p

# Main function
def main():
    # Both persons agree upon the public keys G and P
    # A prime number P is taken
    P = 23
    print("The value of P:", P)

    # A primitive root for P, G is taken
    G = 9
    print("The value of G:", G)

    # Alice chooses the private key a
    # a is the chosen private key
    a = 4
    print("The private key a for Alice:", a)

    # Gets the generated key
    x = power(G, a, P)

    # Bob chooses the private key b
    # b is the chosen private key
    b = 3
    print("The private key b for Bob:", b)

    # Gets the generated key
    y = power(G, b, P)

    # Generating the secret key after the exchange of keys
    ka = power(y, a, P)  # Secret key for Alice
    kb = power(x, b, P)  # Secret key for Bob

    print("Secret key for Alice is:", ka)
    print("Secret key for Bob is:", kb)

if __name__ == "__main__":
    main()
```

```python
import random

# Function to perform modular exponentiation
# It returns (x^y) % p

def power(x, y, p):
    res = 1  # Initialize result
    x = x % p  # Update x if it is more than or equal to p
    while y > 0:
        # If y is odd, multiply x with result
        if y & 1:
            res = (res * x) % p
        # y must be even now
        y = y >> 1  # y = y/2
        x = (x * x) % p  # Change x to x^2
    return res

# Driver code
if __name__ == '__main__':
    # Both the parties agree upon the public keys G and P
    P = 23  # A prime number P is taken
    G = 9   # A primitive root for P, G is taken

    print("The Value of P:", P)
    print("The Value of G:", G)

    # Alice will choose the private key a
    a = random.randint(1, P-1)
    print("The Private Key a for Alice:", a)

    # Bob will choose the private key b
    b = random.randint(1, P-1)
    print("The Private Key b for Bob:", b)

    # Alice will send Bob A = G^a mod P
    A = power(G, a, P)
    print("The Public Key A from Alice to Bob (G^a mod P):", A)

    # Bob will send Alice B = G^b mod P
    B = power(G, b, P)
    print("The Public Key B from Bob to Alice (G^b mod P):", B)

    # Alice calculates the shared secret key
    # s1 = B^a mod P
    s1 = power(B, a, P)
    print("Secret key for Alice (B^a mod P):", s1)

    # Bob calculates the shared secret key
    # s2 = A^b mod P
    s2 = power(A, b, P)
    print("Secret key for Bob (A^b mod P):", s2)

    # Both s1 and s2 must be equal
    assert s1 == s2
    print("Shared Secret Key is successfully exchanged and it is:", s1)
```
